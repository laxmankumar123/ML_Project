```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots()

# Draw a rectangle: Rectangle((x, y), width, height)
rect = patches.Rectangle((1, 1), 4, 2, linewidth=2, edgecolor='red', facecolor='blue', alpha=0.3)
ax.add_patch(rect)

# Set limits to see the rectangle
ax.set_xlim(0, 6)
ax.set_ylim(0, 5)

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

# Load an image
img = cv2.imread('moon.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Draw rectangle: (image, top-left, bottom-right, color(BGR), thickness)
cv2.rectangle(img, (30, 30), (200, 150), (255, 0, 0), 3)

plt.imshow(img)
plt.axis('off')
plt.show()
```
